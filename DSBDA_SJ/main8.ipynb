```python
data['sex'] = data['sex'].replace(['male', 'female'], [0, 1])
data['alive'] = data['alive'].replace(['no', 'yes'], [0, 1])
data['alone'] = data['alone'].replace(['False', 'True'], [0, 1])
data['class'] = data['class'].replace(['First', 'Second' , 'Third'], [1, 2 , 3])
```

```python
data = data[['survived', 'pclass', 'sex', 'age','class']]
```
