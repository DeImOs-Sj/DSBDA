```python
import numpy as np 
import pandas as pd
import matplotlib .pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
data
```

```python
sns.histplot(x='sepal_length' , data=data , bins=10 , kde=True)
```

```python
sns.boxplot(x='petal_length',y='species',data=data)
```

```python
import matplotlib.pyplot as plt
f, a = plt.subplots(2, 2, figsize = (16, 9))
sns.histplot(data['sepal_length'], ax = a[0, 0])
sns.histplot(data['sepal_width'], ax = a[0, 1])
sns.histplot(data['petal_length'], ax = a[1, 0])
sns.histplot(data['petal_width'], ax = a[1, 1])
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 2, figsize = (16, 9))
sns.boxplot(y = 'petal_length', x = 'species', data = data, ax = axes[0,0])
sns.boxplot(y = 'petal_width', x = 'species', data = data, ax = axes[0,1])
sns.boxplot(y = 'sepal_length', x = 'species', data = data, ax = axes[1,0])
sns.boxplot(y = 'sepal_width', x = 'species', data = data, ax = axes[1,1])
```
